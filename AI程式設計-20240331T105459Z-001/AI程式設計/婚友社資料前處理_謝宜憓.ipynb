###讀入資料並印出

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/cnchi/datasets/master/Match_Making.csv')
```

```python
import numpy as np
```

```python
print(df)
```

###切分自、應變數

```python
x = df.iloc[:, 1:5].values
y = df.iloc[:, 5].values
```

```python
x
```

```python
y
```

###類別資料數位化

```python
ary_dummies = pd.get_dummies(x[:, 3]).values
X = np.concatenate((ary_dummies, x[:, 0:3]), axis=1).astype("float64")
X
```

```python
from sklearn.preprocessing import LabelEncoder

labelEncoder = LabelEncoder()
Y = labelEncoder.fit_transform(y).astype("float64")
Y
```

###切分訓練、測試集並印出

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
print("X_train:\n", X_train)
print("X_test:\n", X_test)
print("Y_train:\n", Y_train)
print("Y_test:\n", Y_test)
```

###特徵縮放並印出

```python
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler().fit(X_train)
X_train = sc_X.transform(X_train)
X_test = sc_X.transform(X_test)
```
