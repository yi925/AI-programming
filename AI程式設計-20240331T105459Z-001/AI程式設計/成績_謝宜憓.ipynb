###讀入資料並印出

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Student_Score.csv")
print(df)
```

```python
sum(df.isnull().sum())
```

###計算各科平均分

```python
df_math = df.iloc[:, 1].values
df_eng = df.iloc[:, 2].values
df_sci = df.iloc[:, 3].values
```

```python
print("Math Average:",df_math.mean())
print("English Average:",df_eng.mean())
print("Science Average:",df_sci.mean())
```

###繪圖

```python
plt.plot(["Math", "English", "Science"],[80.25, 79.75, 80.7], "-o")
plt.xlabel("Subjects")
plt.ylabel("Average Score")
plt.title("Average Scores in Each Subject")

plt.show()
```
