```python
X = np.arange(1.,31., 1)
Y = df.loc[:,["氣溫(℃)"]]

plt.title("2023年6月溫度折線圖",fontproperties=myfont)

plt.xlabel("日期",fontproperties=myfont)
plt.ylabel("氣溫(℃)",fontproperties=myfont)

plt.plot(X, Y, "b", label="溫度")
plt.legend(loc='upper right', prop=myfont)

plt.show()
```
