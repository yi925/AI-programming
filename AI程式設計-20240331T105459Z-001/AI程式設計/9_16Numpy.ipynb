```python
import numpy as np
```

```python
ary1 = np.array([1,2,3])
ary2 = np.array([[1,2,3],[4,5,6]])

print(ary1)
print(ary2)
```

```python
ary3 = np.array([11, True, "joke"])
print(ary3)
```

```python
zero1 = np.zeros((3,))
zero2 = np.zeros((2,3))

print(zero1)
print(zero2)
```

```python
identity1 = np.eye(1)
identity2 = np.eye(2)

print(identity1)
print(identity2)
```

```python
one1 = np.ones((3,))
one2 = np.ones((2,3))

print(one1)
print(one2)
```

```python
const1 = np.full((3,),7)
const2 = np.full((2,3),7)

print(const1)
print(const2)
```

```python
ary1 = np.array([[1,2,3],[4,5,6]])
print(type(ary1),ary1.ndim,ary1.shape,ary1.dtype)
```

```python
sample1 = np.arange(0.,5.,0.2)
print(sample1)

np.random.shuffle(sample1)
print(sample1)

sample1 = sample1.reshape(5,5)
print(sample1)

sample1 = sample1.astype("unicode")
print(sample1)
```

```python
sample1 = np.linspace(0. ,5. ,5)
print(sample1)

sample2 = np.logspace(0. ,5. ,5)
print(sample2)
```

```python
sample2 = np.random.randint(1,7,size=15)
print(sample2)

sample3 = np.random.rand(2,3)
print(sample3)
```

```python
normal1 = np.random.randn(3,5)
print(normal1)

normal2 = np.random.normal(10,2,size=(3,5))
print(normal2)
```

```python
X1 = np.array([1, 2, 3])
X2 = np.array([20, 36, 40])
features = np.concatenate((X1,X2)).reshape(2,3).T

print(features)
```
