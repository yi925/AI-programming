# 環境設定

```python
# 引入必要套件
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
```

# 折線圖（Line Charts）

```python
# X 軸資料製作
X = np.arange(0., 5., 0.2)

# 繪製三組折線圖
X = np.arange(0., 5., 0.2)
plt.plot(X, X, "r--")
plt.plot(X, X**2, "bs")
plt.plot(X, X**3, "g^")
plt.show()
```

## 解決中文字形問題

```python
# 先檢查 Colab 目前 Python 版本（至小數點下兩位），得知將來字形檔安裝路徑
!python --version

# 下載台北思源黑體，並命名taipei_sans_tc_beta.ttf
!wget -O taipei_sans_tc_beta.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download

# 移至指定路徑（若 Python 版本並非 3.10，則需配合更改）
!mv taipei_sans_tc_beta.ttf /usr/local/lib/python3.10/dist-packages/matplotlib/mpl-data/fonts/ttf

# 自定義字體變數（若 Python 版本並非 3.10，則需配合更改）
myfont = FontProperties(fname=r'/usr/local/lib/python3.10/dist-packages/matplotlib/mpl-data/fonts/ttf/taipei_sans_tc_beta.ttf')

# 後續在相關函式中，使用 myfont 做為指定字形即可。
# 如：plt.xlabel("時間", fontproperties=myfont)
# 如：plt.legend(loc='best', prop=myfont)
```

```python
# X 軸資料製作
X = np.arange(0., 5., 0.2)

# 設定圖形標題
plt.title("X, X平方, X三次方線圖", fontproperties=myfont)

# 設定軸線標籤
plt.xlabel("X 軸", fontproperties=myfont)
plt.ylabel("Y 軸", fontproperties=myfont)

# 繪製三組折線圖並設定圖例
plt.plot(X, X, "r--", label="X")
plt.plot(X, X**2, "bs", label="X平方")
plt.plot(X, X**3, "g^", label="X立方")
plt.legend(loc='best', prop=myfont)

plt.show()
```

# 點狀圖（Scatter Charts）

```python
# 設定點狀圖樣本點數量
n = 150

# 第一群：以 (-1, 2) 為中心，呈常態分佈散開
X1 = np.random.normal(-1, 1, n)
Y1 = np.random.normal(2, 1, n)

# 第二群：以 (2, -1) 為中心，呈常態分佈散開
X2 = np.random.normal(2, 1, n)
Y2 = np.random.normal(-1, 1, n)

# 繪製點狀圖：s=點的大小、c=點的顏色、marker=點的代表符號
plt.scatter(X1, Y1, s=75, c="red", marker="+")
plt.scatter(X2, Y2, s=75, c="blue", marker="o")

plt.show()
```
