```python
from tensorflow.keras.datasets import mnist

# Load Database
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
import HappyML.model_drawer as md

md.show_first_n_images(x_ary=x_train, y_real=y_train, first_n=10)
```

```python
base_dir = 'cats_and_dogs_filtered'        #資料集標題
train_dir = os.path.join(base_dir, 'train')   #訓練集
test_dir = os.path.join(base_dir, 'validation') #測試集
```

```python
# Generate Training / Validation Set
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1.0/255,       #縮放至0~1
    validation_split=0.25,    #訓練75%驗證25%
    shear_range=10,       #隨機切變、縮放、翻轉
    zoom_range=0.2,
    horizontal_flip=True)

train_set = train_datagen.flow_from_directory(  #訓練集
    directory=train_dir,
    target_size=(64, 64),
    batch_size=10,
    class_mode="binary",
    subset="training")

val_set = train_datagen.flow_from_directory(  #驗證集
    directory=train_dir,
    target_size=(64, 64),
    batch_size=10,
    class_mode="binary",
    subset="validation")
```
