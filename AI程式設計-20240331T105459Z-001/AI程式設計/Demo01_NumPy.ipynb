# 環境設定

```python
# 引入必要的套件
import numpy as np
```

# 建立陣列

## 建立一般陣列

```python
# 建立一維陣列
ary1 = np.array([1, 2, 3])
print(ary1)

# 建立二維陣列
ary2 = np.array([[1, 2, 3], [4, 5, 6]])
print(ary2)
```

```python
# 於 NumPy 陣列內，建立不同資料型態的元素
ary3 = np.array([15, "Apple", True])   #通通變字串
print(ary3)
```

## 建立特殊陣列

```python
# 建立一維「零陣列」
zero1 = np.zeros((3, ))
print(zero1)

# 建立二維「零陣列」
zero2 = np.zeros((2, 3))
print(zero2)
```

```python
# 建立一維「單位陣列」
identity1 = np.eye(1)
print(identity1)

# 建立二維「單位陣列」
identity2 = np.eye(2)
print(identity2)
```

```python
# 建立全為 1 的一維「常數陣列」
one1 = np.ones((3, ))
print(one1)

# 建立全為 1 的二維「常數陣列」
one2 = np.ones((2, 3))
print(one2)

# 建立全為特定數字（如：7）的一維「常數陣列」
const1 = np.full((3, ), 7)
print(const1)

# 建立全為特定數字（如：7）的二維「常數陣列」
const2 = np.full((2, 3), 7)
print(const2)
```

# 讀寫陣列元素

```python
# 讀取陣列元素
ary1 = np.array([[1, 2, 3], [4, 5, 6]])
print(ary1[0, 0], ary1[1, 2])
```

```python
# 寫入陣列元素
ary1[0, 0] = 100
ary1[1, 2] = 600
print(ary1)
```

# 讀取陣列資訊

```python
# 產生陣列
ary1 = np.array([[1, 2, 3], [4, 5, 6]])

# 讀取陣列各種資訊
print(type(ary1), ary1.ndim, ary1.shape, ary1.dtype)
```

# 產生「樣本點」

## 產生「線性」樣本點

```python
# 產生「線性」樣本點
sample1 = np.arange(0., 5., 0.2)
print(sample1)
```

```python
# 產生「線性」樣本點 + 洗牌
sample1 = np.arange(0., 5., 0.2)
np.random.shuffle(sample1)
print(sample1)
```

```python
# 產生「線性」樣本點 + 更改維度
sample1 = np.arange(0., 5., 0.2)
print(sample1.reshape(5, 5))
```

```python
# 產生「線性」樣本點 + 更改資料型態
sample1 = np.arange(0., 5., 0.2)
sample1 = sample1.astype("unicode")
print(sample1)
```

## 產生「等差 / 等比」樣本點

```python
# 產生「等差樣本點」
sample1 = np.linspace(0., 5., 5)
print(sample1)
```

```python
# 產生「等比樣本點」
sample1 = np.logspace(0., 5., 5)
print(sample1)
```

## 產生「亂數」樣本點

```python
# 產生「整數亂數」樣本點
sample2 = np.random.randint(1, 7, size=15)
print(sample2)
```

```python
# 產生「浮點數亂數」樣本點
sample3 = np.random.rand(2, 3)
print(sample3)
```

## 產生「常態分佈」樣本點

```python
# 產生「標準常態分佈」樣本點
normal1 = np.random.randn(3, 5)
print(normal1)
```

```python
# 產生「一般常態分佈」樣本點
normal2 = np.random.normal(10, 2, size=(3, 5))
print(normal2)
```

# 切片運算（Slicing）

```python
# 一維陣列切片運算
slice1 = np.random.randint(20, size=20)
print(slice1)
print(slice1[:5])
```

```python
# 二維陣列切片運算
slice2 = np.random.randint(20, size=(5, 5))
print(slice2)
print(slice2[:3, :3])
```

# 統計量的計算

```python
# 極小、極大、總和、平均、標準差的計算
stat1 = np.arange(1, 11)
print(stat1)

print(stat1.min())
print(stat1.max())
print(stat1.sum())
print(stat1.mean())
print(stat1.std())
```

# 陣列轉置（Transpose）

```python
# 陣列並排合併後轉置
X1 = np.array([1, 2, 3])
X2 = np.array([20, 36, 40])

features = np.concatenate((X1, X2)).reshape(2, 3).T
print(features)
```
